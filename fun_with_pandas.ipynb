## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

#%pip install lxml
import pandas as pd
import numpy as np

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")

```

```python
# inspecting the data
print(rawdata)
 
# 1) Combines the tables of best-selling physical singles and best-selling digital singles on the 
# Wikipedia page "List_of_best-selling_singles"

# tables to use are: 1 + 2 (phsycial singles) and 4 +5 (digital singles)

physical_and_digital = pd.concat([rawdata[0], rawdata[1], rawdata[3], rawdata[4]], axis=0)
```

```python
# sanity check
list(physical_and_digital)
```

```python
# 2. Adds a column which marks whether each row is from the list of physical singles or digital singles
physical_and_digital['Type'] = ""

# as release year seems to divide the digital and physical singles, im using that to assign the type of single to each row
physical_and_digital['Type'] = np.where(physical_and_digital['Released'] >= 2004, 'Digital', 'Physical')

print(physical_and_digital)
```

```python
# 3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take
# the closest year after you were born.

# counting up from 2001 until a release year matches
year = 2001
while year not in physical_and_digital['Released'].values:
    year += 1  # Increment the year by 1 each time

print(physical_and_digital.loc[physical_and_digital['Released'] == year, ['Artist', 'Single', 'Released']])
    
# I was born in 2001, but no songs were released in that year. The closest year, where a single was released, is 2003, where O-Zone released "Dragostea din tei"
```

```python
# 4. Outputs the artist and single name for the year you were 15 years old.

year = 2001 + 15
while year not in physical_and_digital['Released'].values:
    year += 1  # Increment the year by 1 each time

print(physical_and_digital.loc[physical_and_digital['Released'] == year, ['Artist', 'Single', 'Released']])

```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
```

```python
# 1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
space_missions = pd.concat(rawdata[0:8], axis=0)

print(space_missions)
```

```python
# 2. Write a script that returns the year with the most launches
space_missions['Year'] = space_missions['Launch date'].str[-5:]

year_counts = space_missions['Year'].value_counts()

year_counts.sort_values

year_counts = year_counts.reset_index()

print(year_counts.iloc[:2])
```

```python
# 3. Write a script that returns the most common month for launches
space_missions['Month'] = space_missions['Launch date'].str.extract('([a-zA-Z]+)', expand=False)

month_counts = space_missions['Month'].value_counts()

month_counts.sort_values

month_counts = month_counts.reset_index()

print(month_counts.iloc[0])
```

```python
# 4. Write a script that ranks the months from most launches to fewest launches

print(month_counts)
```

## Supervillain challenge

1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade
3. Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
```

```python
# 1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
supervillains = pd.concat(rawdata[3:12]) # as the first supervillain table is the third table of the page
list(supervillains)
print(supervillains)
```

```python
# making sure i got all the data in the dataframe by comparing the last value with the last value in the table of the webpage
print(supervillains.iloc[626])
```

```python
supervillains['Decade'] = supervillains['Year Debuted'].str.findall('(\d+)')
print(supervillains['Decade'])
```

```python
supervillains['Year'] = supervillains['Year Debuted'].str.extract('(\d{4})').astype(int)
supervillains['Decade'] = supervillains['Year'].apply(lambda year: str(10* (year // 10)) + 's')
```

```python
# 3. Write a script that ranks the different comics companies in terms of how many supervillains they have, 
# and display the results in a nice table (pandas dataframe)

companies_counts = supervillains['Company'].value_counts()

companies_counts_df = companies_counts.reset_index()

# giving the columns names
companies_counts_df.columns = ['Company', 'Count']

print(companies_counts_df)
```

## Stroop challenge

Every year between 2015 and 2021, the students in my Language, Cognition, and the Brain course participated in a version of the Stroop task. Using a stopwatch (ok, using their phones), they recorded how fast they could say a list of things (either reading or naming colors or color words). The column names mean "Reading with No Interference", "Naming with Interference", "Naming with No Interference", and "Reading with Interference". The times are in seconds.

### Stroop challenge 1: 
Transform these data from wide format to long format, so that the result is a dataframe with
- 1 column named "Participant_id" with a unique number for each participant (you can use the row indices)
- 1 column named "Year" with the year data
- 1 column named "Task" that shows which task they were doing
- 1 column named "RT" that shows their response time

```python
# Starter code 1...

df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/Stroop/master/Stroop-raw-over-the-years.csv")
df.head()
```

```python
# Starter code 2...

# Make a new column using the dataframe indices as particpant numbers

df.index.name = 'Participant_id'
df = df.reset_index()

#df.reset_index(inplace = True)
# NOTE: This line does exactly the same thing as the line above:
# it replaces the original df with a new df with the updated index. That's what
# "inplace = True" means. Or, you can just assign the dataframe with the updated index
# to a new dataframe with the same name as the old dataframe, which is what I did above.
# The end result is the same.

df_long = pd.melt(df, id_vars=['Participant_id', 'Year'], var_name='Task', value_name='RT')
```

```python
# checking that there are actually 4 rows for each participant
df_long.sort_values(by= 'Participant_id')
df_long.head
```

## Stroop challenge 2 (Advanced!!!):

Make a new dataframe which shows the mean response time (in seconds) for each task for each year.

```python
df_mean = df_long.groupby(['Year', 'Task'])['RT'].mean()

df_mean = df_mean.reset_index()
print(df_mean)
```
